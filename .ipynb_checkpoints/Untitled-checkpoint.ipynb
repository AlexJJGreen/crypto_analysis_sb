```python
import pandas as pd
import numpy as np
```

```python
# create df
df = pd.read_csv("eur_pairs/BTC-EUR.csv", parse_dates=True)
df.index = df["Date"]
df.drop(["Date"],axis=1,inplace=True)

# add calculations
df["SMA50"] = df["Close"].rolling(50).mean()
df["SMA200"] = df["Close"].rolling(200).mean()
df["chg%"] = (df["Close"] / df["Close"].shift(1)) - 1

# abv 50 abv 200
abv50_abv200 = df[(df["Close"] > df["SMA50"]) & (df["Close"] > df["SMA200"])]
abv50 = df[df["Close"] > df["SMA50"]]
```
