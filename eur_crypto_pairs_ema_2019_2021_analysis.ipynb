```python
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join
```

```python
DIR_PATH = "eur_pairs/"
FILES = [f for f in listdir(dir_path)]
OUTPUTS = []
```

```python
def col_total(df):
    return df[df.columns[0]].count()

def output_gen(df, output_list):
    return output_list.append(df.name), output_list.append({df.name + "_len": col_total(df)}), output_list.append({df.name + "_chg": df["chg%"].sum()}), output_list.append({df.name + "_mean": df["chg%"].mean()})
```

```python
for current_file in FILES:
    # create df
    df = pd.read_csv(DIR_PATH + current_file, parse_dates=True)
    df.name = current_file
    df.index = df["Date"]
    df.drop(["Date"],axis=1,inplace=True)
    
    # add calculations
    df["sma9"] = df["Close"].rolling(9).mean()
    df["sma20"] = df["Close"].rolling(20).mean()
    df["sma50"] = df["Close"].rolling(50).mean()
    df["chg%"] = (df["Close"] / df["Close"].shift(1)) - 1
    
    # get data set as list of dicts and append to output
    output = []
    output.append(current_file)
    # main
    output.append({"total_len": col_total(df)})
    output.append({"total_chg": df["chg%"].sum()})
    output.append({"total_mean": df["chg%"].mean()})
    # abv 9 abv 20 abv 50
    abv9_abv20_abv50 = df[(df["Close"] > df["sma9"]) & (df["Close"] > df["sma20"]) & (df["Close"] > df["sma50"])]
    abv9_abv20_abv50.name = "abv9_abv20_abv50"
    output_gen(abv9_abv20_abv50, output)
    # abv 9 abv 20 blw 50
    abv9_abv20_blw50 = df[(df["Close"] > df["sma9"]) & (df["Close"] > df["sma20"]) & (df["Close"] < df["sma50"])]
    abv9_abv20_blw50.name = "abv9_abv20_blw50"
    output_gen(abv9_abv20_blw50, output)
    # abv 9 blw 20 blw 50
    abv9_blw20_blw50 = df[(df["Close"] > df["sma9"]) & (df["Close"] < df["sma20"]) & (df["Close"] < df["sma50"])]
    abv9_blw20_blw50.name = "abv9_blw20_blw50"
    output_gen(abv9_blw20_blw50, output)
    # abv 9 blw 20 abv 50
    abv9_blw20_abv50 = df[(df["Close"] > df["sma9"]) & (df["Close"] < df["sma20"]) & (df["Close"] > df["sma50"])]
    abv9_blw20_abv50.name = "abv9_blw20_abv50"
    output_gen(abv9_blw20_abv50, output)
    # blw 9 blw 20 abv 50
    blw9_blw20_abv50 = df[(df["Close"] < df["sma9"]) & (df["Close"] < df["sma20"]) & (df["Close"] > df["sma50"])]
    blw9_blw20_abv50.name = "blw9_blw20_abv50"
    output_gen(blw9_blw20_abv50, output)
    # blw 9 abv 20 abv 50
    blw9_abv20_abv50 = df[(df["Close"] < df["sma9"]) & (df["Close"] > df["sma20"]) & (df["Close"] > df["sma50"])]
    blw9_abv20_abv50.name = "blw9_abv20_abv50"
    output_gen(blw9_abv20_abv50, output)
    # blw 9 abv 20 blw 50
    blw9_abv20_blw50 = df[(df["Close"] < df["sma9"]) & (df["Close"] > df["sma20"]) & (df["Close"] < df["sma50"])]
    blw9_abv20_blw50.name = "blw9_abv20_blw50"
    output_gen(blw9_abv20_blw50, output)
    # blw 9 blw 20 blw 50
    blw9_blw20_blw50 = df[(df["Close"] < df["sma9"]) & (df["Close"] < df["sma20"]) & (df["Close"] < df["sma50"])]
    blw9_blw20_blw50.name = "blw9_blw20_blw50"
    output_gen(blw9_blw20_blw50, output)
    OUTPUTS.append(output)
```
